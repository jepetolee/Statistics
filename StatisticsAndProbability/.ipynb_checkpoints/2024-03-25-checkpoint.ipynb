## 공분산: 2개의 연속형 변수의 관계를 볼 때 사용 되는 통계

```python
import pandas as pd
import numpy as np
%precision 3 # "%.3f"
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
cov_data = pd.read_csv("3-2-3-cov.csv")
print(cov_data)
```

```python
# 데이터 분리
x = cov_data["x"]
y = cov_data["y"]
# sample size
N = len(cov_data)
# 평균값 계산
mu_x = np.mean(x)
mu_y = np.mean(y)
```

```python
cov_sample = sum((x - mu_x) * (y - mu_y)) / N
cov_sample
```

```python
cov = sum((x - mu_x) * (y - mu_y)) / (N - 1)
cov
```

```python
np.cov(x,y,ddof=0)
```

```python
np.cov(x,y,ddof=1)
```

```python
sigma_2_x  = np.var(x,ddof =1)
sigma_2_y  = np.var(y,ddof =1)
rho = cov/ np.sqrt(sigma_2_x*sigma_2_y)
rho
```

```python
sigma_2_x_sample  = np.var(x,ddof =0)
sigma_2_y_sample  = np.var(y,ddof =0)
rho = cov_sample/ np.sqrt(sigma_2_x_sample*sigma_2_y_sample)
rho
```

## 히스토그램 평활화
### 구간을 나누는 것이 아니라 히스토그램을 부드러운 선으로 연결
#### 커널 밀도 추정 (Kernel Density Estimation)
#### 히스토그램 평활화를 위한 한가지 방법.





```python
fish_multi = pd.read_csv("3-3-2-fish_multi_2.csv")
fish_multi.groupby("species").describe()
```

```python
# 데이터를 물고기 종류에 따라 나눔 !-- 시험문제
length_a = fish_multi.query('species == "A"')["length"]
length_b = fish_multi.query('species == "B"')["length"] 
```

## 히스토그램을 그리는 함수를 2 회 연속으로 실행

```python
sns.distplot(length_a, bins = 5, color = 'red', kde = False)
sns.distplot(length_b, bins = 5, color = 'blue', kde = False)
```

```python
sns.displot(length_a, bins = 5, color = 'red', kde = False) # 이친구들은 두개가 연속으로 되진 않음.
sns.displot(length_b, bins = 5, color = 'blue', kde = False)
```

```python
sns.histplot(length_a, bins = 5, color = 'red', kde = False)
sns.histplot(length_b, bins = 5, color = 'blue', kde = False)
```
